```python
import numpy as np
import pandas as pd
import sep
import fitsio
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib.patches import Ellipse
%matplotlib inline
import glob

rcParams['figure.figsize'] = [10., 8.]
```

```python
targetx = 1836
targety = 720
name = '003325-5-0049'

colour = 'z'
data = fitsio.read('frame-' + colour + '-' + name + '.fits')
```

```python
# colour g
bkg = sep.Background(data)
bkg_image = bkg.back()
data_sub = data
objects = sep.extract(data_sub, 1.5, err=bkg.globalrms)

flux, fluxerr, flag = sep.sum_circle(data_sub, objects['x'], objects['y'],
                                     3.0, err=bkg.globalrms, gain=1.0)

np.savetxt('target-' + colour + '.csv',objects ,delimiter=',')

names = ['thresh','npix','tnpix','xmin','xmax','ymin','ymax','x','y','x2','y2','xy','errx2','erry2','errxy','a','b','theta','cxx','cyy','cxy','cflux','flux','cpeak','peak','xcpeak','ycpeak','xpeak','ypeak','flag']
tin = pd.read_csv ('target-' + colour + '.csv', names=names)

targetdata = tin
targetdata2 = tin[((tin.x < targetx+20) & (tin.x > targetx-20) & (tin.y < targety+20) & (tin.y > targety-20))]
targetdata

targetdata.to_csv('target-' + colour + '.csv')
targetdata2.to_csv('targetdata-' + colour + '.csv')
```
